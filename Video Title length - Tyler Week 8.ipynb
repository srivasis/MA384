# Video Title Length

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import CleanDataFun as clean
import Merge_Country_Data as merge
```

```python
usData = clean.cleanData('../Data/Youtube/USvideos.csv', '../Data/Youtube/US_category_id.json', "latin_1")
usData['category'] = usData['category'].astype('category')
gbData=clean.cleanData('../Data/Youtube/GBvideos.csv', '../Data/Youtube/GB_category_id.json', "latin_1")
gbData['category'] = gbData['category'].astype('category')
allCountries = merge.allParcableCountryData()
allCountries['category'] = allCountries['category'].astype('category')
allCountries['country'] = allCountries['country'].astype('category')
```

# Video Title Length

## US

```python
VideoTitleLengthUS = pd.DataFrame()
#VideoTitleLength['title'] = usData['title']
VideoTitleLengthUS['views'] = usData['views']
VideoTitleLengthUS['length'] = usData['title'].str.len()
VideoTitleLengthUS.head(3)
```

```python
VideoTitleLengthUS.sort_values('length').plot.line(x='length',y='views')
```

```python
plt.figure(figsize=(16, 6))
ax = sns.regplot(x="length", y="views", data=VideoTitleLengthUS, x_estimator=np.mean, logx=True, truncate=True)
ax.set_title('US')
plt.show()
```

## Great Britain

```python
VideoTitleLengthGB = pd.DataFrame()
VideoTitleLengthGB['views'] = gbData['views']
VideoTitleLengthGB['length'] = gbData['title'].str.len()
VideoTitleLengthGB.head(3)
```

## Combined country data

```python
VideoTitleLengthAll = pd.DataFrame()
VideoTitleLengthAll['views'] = allCountries['views']
VideoTitleLengthAll['length'] = allCountries['title'].str.len()
VideoTitleLengthAll.head(3)
```

```python
VideoTitleLengthAll.sort_values('length').plot.line(x='length',y='views',title="Avg Views vs. Title length",figsize=(15,10))
```

```python
plt.figure(figsize=(16, 6))
ax = sns.regplot(x="length", y="views", data=VideoTitleLengthAll, x_estimator=np.mean, logx=True, truncate=True)
ax.set_title('All Countries - Views vs. Title length')
plt.show()
```
