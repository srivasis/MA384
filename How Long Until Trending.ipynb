```python
usData['publish_time'] = pd.to_datetime(usData['publish_time'],format='%Y-%m-%d', errors='coerce')
usData['trending_date'] = pd.to_datetime(usData['trending_date'],format='%Y-%m-%d', errors='coerce')
```

```python
data = usData[usData.publish_time.notnull()]
data = data[data.trending_date.notnull()]
data.head()
```

```python
data.sort_values(by='publish_time', ascending = False)
```

```python

data['dif'] = data['trending_date']- data['publish_time']
#data['dif'] = data['dif'].apply(days, axis=1)
data['dif'] = data['dif'].dt.days
data.head()
data['dif'].plot(title ="Top 10 Most Disliked Video")
data.sort_values(by='dif', ascending = False)
```

```python
dataGrouped = data.groupby(data.publish_time)
cleaned = data.sort_values('dif', ascending=False).drop_duplicates('trending_date')
#cleaned.drop_duplicates('title').sort_values('dif', ascending=False)
```

In February 2018, it appears that videos that where published over 36 days ago no longer make it to trending. Changes in YouTube's algorithm for trending videos may have been a result of scrutiny from viewers after Logan Paul released a video featuring the aftermath of a suicide on the platform a month earlier. Major news sites like CNBC and the Guardian wrote articles critisizing YouTube's trending algorithm for helping this type of material surface (https://www.cnbc.com/2018/02/22/youtube-trending-algorithm-needs-reform.html, https://www.theguardian.com/technology/2018/feb/02/how-youtubes-algorithm-distorts-truth).
