```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import CleanDataFun as clean
import Merge_Country_Data as merge
from sklearn.decomposition import TruncatedSVD
```

```python
usData = clean.cleanData('../Data/Youtube/USvideos.csv', '../Data/Youtube/US_category_id.json', "latin_1")
usData['category'] = usData['category'].astype('category')
gbData=clean.cleanData('../Data/Youtube/GBvideos.csv', '../Data/Youtube/GB_category_id.json', "latin_1")
gbData['category'] = gbData['category'].astype('category')
allCountries = merge.allParcableCountryData()
allCountries['category'] = allCountries['category'].astype('category')
allCountries['country'] = allCountries['country'].astype('category')
```

```python
allCountries.head(3)
```

```python
allCountries_number_attributes = allCountries[['views','likes','dislikes','comment_count']]
allCountries_number_attributes
```

```python
# can see a fairly high correlation between many of the attributes besides dislikes.
allCountries_number_attributes.corr()
```

## Compressing views,likes,dislikes, & comment count into 3 components

```python
svd = TruncatedSVD(n_components=3)
scores_compressed = pd.DataFrame(svd.fit_transform(allCountries_number_attributes))
scores_compressed
```

```python
percentExplained = pd.Series(svd.explained_variance_ratio_)
percentExplained
```

```python
# Looks as through the data could be represented by 1 compressed component
```

## Compressed to 1 component

```python
svd = TruncatedSVD(n_components=1)
scores_compressed = pd.DataFrame(svd.fit_transform(allCountries_number_attributes))
```

```python
print('Percentage of data explained by 1 compressed component:')
print((100 * svd.explained_variance_ratio_[0]).round(2))
```
